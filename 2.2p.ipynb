Уточните методом половинного деления наименьший по модулю и отличный от нуля корень уравнения xsinx-1=0 с точностью до 0.00001

```python
from math import sin
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Определение функций
def f(x):
    return x * sin(x) - 1

def y1(x):
    return x * sin(x) 

def y2(x):
    return 1
```

```python
# СОздание графика
x_vals = np.linspace(-10, 10, 500)
y1_vals = [y1(x) for x in x_vals]
y2_vals = [y2(x) for x in x_vals]
fig, ax = plt.subplots()
ax.plot(x_vals, y1_vals, color='green', label='y=xsinx')
ax.plot(x_vals, y2_vals, color='blue', label='y=1')
ax.xaxis.set_major_locator(ticker.MultipleLocator(1)) 
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
plt.grid(True)
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.title('График функций')
```

```python
# Нахождение интервалов, внутри которых существует корень
intervals = []
step = 0.1
x = -10
while x < 10:
    if f(x) * f(x + step) < 0:
        intervals.append((x, x + step))
    x += step
```

```python
# Уточнение корня на каждом интервале
roots = []
for interval in intervals:
    x1, x2 = interval
    eps = 0.00001
    while abs(x2 - x1) > eps:
        c = (x1 + x2) / 2
        if f(x1) * f(c) <= 0:
            x2 = c
        else:
            x1 = c
    root = (x1 + x2) / 2
    roots.append(root)
```

```python
y_roots = [y1(r) for r in roots]
ax.plot(roots, y_roots, 'ro', label='Корни')  
```

```python
for i in range(len(intervals)):
        print(f'Интервал №{i+1} = {intervals[i]}, корень на нем с ошибкой менее {eps} равен = {roots[i]}')

plt.show()
```
